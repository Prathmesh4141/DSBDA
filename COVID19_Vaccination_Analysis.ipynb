## 4. Handling Missing Values

### 4.1 Fill First Dose with Column Mean

```python
mean_first = df["First Dose Administered"].mean()
df["First Dose Administered"].fillna(mean_first, inplace=True)
print("Mean used for First Dose:", round(mean_first, 2))
```

### 4.2 Fill Second Dose with Column Mean

```python
mean_second = df["Second Dose Administered"].mean()
df["Second Dose Administered"].fillna(mean_second, inplace=True)
print("Mean used for Second Dose:", round(mean_second, 2))
```

### 4.3 Confirm No Missing Values Remain in Dose Columns

```python
print(df[["First Dose Administered", "Second Dose Administered"]].isnull().sum())
```

## 5. Data Analysis

### 5.1 State-wise Total First Dose Count

```python
first_dose_state = df.groupby('State')['First Dose Administered'].sum().reset_index()
first_dose_state.columns = ['State', 'Total First Dose']
first_dose_state.sort_values('Total First Dose', ascending=False)
```

### 5.2 State-wise Total Second Dose Count

```python
second_dose_state = df.groupby('State')['Second Dose Administered'].sum().reset_index()
second_dose_state.columns = ['State', 'Total Second Dose']
second_dose_state.sort_values('Total Second Dose', ascending=False)
```

### 5.3 Total Male Individuals Vaccinated

```python
total_male = int(df["Male(Individuals Vaccinated)"].sum())
print(f"Total Male Vaccinated: {total_male:,}")
```

### 5.4 Total Female Individuals Vaccinated

```python
total_female = int(df["Female(Individuals Vaccinated)"].sum())
print(f"Total Female Vaccinated: {total_female:,}")
```
