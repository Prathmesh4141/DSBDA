```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("TSLA_Stock.csv")
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df['Year'] = df['Date'].dt.year 
```

```python
df.head()
```

```python
df.groupby('Year').mean()
```

```python
df.groupby('Year').median()
```

```python
df.groupby('Year').min()
```

```python
df.groupby('Year').max()
```

```python
df.groupby('Year').std()
```

```python
df['Month'] = df['Date'].dt.month
df.groupby('Month').mean()
```
